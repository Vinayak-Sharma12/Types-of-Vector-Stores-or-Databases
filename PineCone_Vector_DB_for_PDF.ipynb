```python
from google.colab import userdata
api_key=userdata.get('pinecone_api_key')
```

## **LOAD DOCUMENT**

```python
from langchain.document_loaders import PyPDFLoader
loader=PyPDFLoader("/content/ncert_agriculture.pdf")
docs=loader.load()
```

```python
len(docs)
```

## **TEXT SPLITTER**

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
chunks = splitter.split_documents(docs)
len(chunks)
```

```python
chunks[0].page_content
```

## **DEFINE EMBEDDING MODEL**

```python
from langchain.embeddings import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

## **CREATE PINECONE VECTOR DB**

### **CREATE CLIENT**

```python
from pinecone import Pinecone
pc=Pinecone(api_key=api_key)
```

### **CREATE INDEX**

```python
from pinecone import ServerlessSpec
index_name='newest-index'
if not pc.has_index(index_name):
    pc.create_index(
         name=index_name,
         dimension=384, #Whatever the dimension is of your embedding
         metric='cosine',
         spec=ServerlessSpec(cloud="aws",region='us-east-1'),
          )
index=pc.Index(index_name)
```

```python
from langchain_pinecone import PineconeVectorStore
vector_db=PineconeVectorStore(index=index,embedding=embeddings)
```

## **ADD DOCUMENTS TO VECTOR DB**

```python
vector_db.add_documents(chunks)
```

## **QUERY TO THE VECTOR DATABASE**

```python
results=vector_db.similarity_search(query="Tell me about farming techniques")
```

```python
print(results[0].page_content)
```
